**LOAD CSV FILE**

This reads the CSV file into a pandas DataFrame called df and cleans up any stray spaces in the column names.

```python
import pandas as pd

df = pd.read_csv("sample_sales_data.csv")
df.columns = [c.strip() for c in df.columns]
print(df.shape)
df.head()
```

**INSPECT ROWS**

This checks its shape, first/last rows, data types, and overall structure to understand what you're working with.


```python
df.tail()
```

```python
df.dtypes
```

```python
df.info()
```

**SUMMARY STATS**

This calculates key stats such as mean, median, min, max, and counts for the numeric columns, plus grouped averages and category counts.

```python
df.describe(include ='all')
```

```python
print(df["Price"].mean() ,
      df["Price"].median(),
      df["Price"].min(),
      df["Price"].max(),
      df["Price"].count())
```

```python
df.groupby("Category")["Price"].mean().round(2)
```

```python
df["Region"].value_counts(dropna=False)
```

**FILTER/SELECTION/SLICE**

This narrows the dataset down to specific rows and columns of interest, such as high-value orders or a particular region and category.


```python
high_value = df[df["Price"] > 500]
high_value[["Product", "Price", "Region"]].head()
```

```python
subset_cols = df[["Product", "Category", "Price"]]
subset_cols.head()
```

```python
df.iloc[10:15]
```

**SAVE FILTERED RESULTS**

This writes the filtered data out to both a CSV file and an Excel file so the results are stored.

```python
high_value.to_csv("high_value_orders.csv", index=False)
high_value.to_excel("high_value_orders.xlsx", index=False)
print("Saved high_value_orders.csv and high_value_orders.xlsx")
```
